```python
# =========================================
# Step: Converting numeric columns safely
# =========================================

# These are the main numeric columns we will use
num_cols = ['CO(GT)', 'C6H6(GT)', 'Temperature', 'Relative Humidity', 'Absolute Humidity']

# Basic loop with existence check so no KeyError happens
for col in num_cols:
    if col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].str.replace(',', '.')
        df[col] = pd.to_numeric(df[col], errors='coerce')
    else:
        # simple student-style placeholder variable, no print
        _ = f"{col} not found, skipping"

# small check to make sure all numeric columns are handled
_ = df[num_cols].dtypes
```
