```python
reshaped = pd.melt(df,
                   id_vars=['Type', 'Category'],
                   value_vars=['comment', 'like', 'share'],
                   var_name='Metric',
                   value_name='Count')
reshaped.head()
```

```python
pivoted = reshaped.pivot_table(index=['Type', 'Category'],
                               columns='Metric',
                               values='Count',
                               aggfunc='mean')
pivoted.head()
```
